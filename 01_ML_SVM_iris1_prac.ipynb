### 1. Import module

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns# seaborn 그래프 라이브러리
import sklearn.datasets
from sklearn.svm import SVC # SVC 함수의 SVM 라이브 러리 
from sklearn.model_selection import train_test_split # 트레이닝과 테스트의 데이터또는 컬럼을 나눠주는 라이브러리
from sklearn.metrics import accuracy_score # 머신러닝에서 결과점수를 연산해주는 라이브러리

import warnings# 에러를 제거하는 라이브러리
warnings.filterwarnings('ignore')
```

### 2. Load Data

```python
iris = sklearn.datasets.load_iris() # sklearn.datasets에 저장된 데이터 불러옴
```

```python
# 데이터를 출력해보면 아래와 같이 2중배열로 저장되있다.
# data, target, frame, target_names, DESCR, feature_names, filename 의 딕셔너리형태의
# 키 값으로 저장되있는 것을 알 수 있다.
iris 
```

```python
iris.data 
```

```python
iris.feature_names
```

```python
# df의 변수에 DataFrame형태로 저장한다.
# 이때 데이터는 iris.data로 넣고 컬럼은 iris.feature_names 저장한다.
# 이유는 data와 feature_names의 컬럼갯수가 동일하기 때문.
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# 데이터를 출력해보면 아래 결과와 같이 각 컬럼에 해당하는 데이터가
# 들어간것을 볼 수 있다.
df.head(1)
```

```python
# iris에 target의 키값이 있는데 target에 어떤 값이 있는지 확인 한다.
# 확인 결과 0 ~ 2의 값이 들어가 있다.

iris.target
```

```python
# target이라는 컬럼을 만들어 주고 그 안에 해당 값들을 넣어준다.
df['target'] = iris.target
df.head()
```

```python
# map을 사용해서 target컬럼의 데이터를 보기좋게 바꿔준다.
# 이때 map이라는 함수를 사용하는데 딕셔너리 형태로
# 컬럼의 벨류값을 지정해서 아래와같이 문자열로 넣어주면 해당 벨류값에 맞게
# 벨류값이 변경되는 것을 알 수 있다.
df['target'] = df['target'].map({0:'setosa',1:'versicolor',2:'virginica'})
df
```

```python
# 컬럼값을 인덱스기준 0에서 n-1까지 출력해준다. 
df.iloc[:,:4].head()
```

```python
# 컬럼명기준 sepal length (cm)에서 petal width (cm)까지 출력해준다. 
df.loc[:,'sepal length (cm)':'petal width (cm)'].head()
```

```python
# 컬럼값을 인덱스기준으로 4번째의컬럼의 전체를 가져옴
df.iloc[:,4].head() 
```

```python
# 컬럼명에 해당하는 컬럼의 전체 데이터를 가져옴
df.loc[:,'target'].head()
```

### 3. Data Colection

```python
# train_test_split() 함수를 사용해서 train데이터와 test데이터를 각각 나눠준다.
# test_size=0.1 으로 설정 하므로써,
# X_train에는 각 속성들의 데이터를 0.9만큼 저장
# X_test 에는 각 속성들의 데이터를 0.1만큼 저장.
# y_train 에는 라벨의 데이터를 0.9만큼 저장
# y_test 에는 라벨의 데이터를 0.1만큼 저장
# shuffle은 데이터들의 섞임을 지정해줄수 있다.

X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :4], df.iloc[:,4],test_size=0.1,shuffle=False ) 
```

각각의 데이터가 변수로 저장이되고,  
train에서는 90%, test에서는 10%의 데이터만 저장된것을 확인 할 수있다.  

```python
X_train.tail(1)
```

```python
X_test.tail(1)
```

```python
y_train.tail(1)
```

```python
y_test.tail(1)
```

### 4. Model Generate and Training

```python
# 모델의 SVC()함수를 model변수에 저장한다.
model = SVC() 

# 저장된 변수에 fit()함수를 사용해서 (데이터=라벨)을 학습 시켜준다.
model.fit(X_train, y_train) 
```

### 5. Prediction

```python
# predict의 변수에 학습된 모델에 임의의 값을 넣어서 연산된 결과를 저장 시킨다.
predict = model.predict(X_test)
print(predict)
```

```python
# 데이터가 얼마나 정확 했는지 확인해 봅니다.
# y_test 데이터 인덱스번호가 135부터 있는걸 감안해 더해줍니다.
sum = 0
for i in range(len(predict)):
    if predict[i] == y_test[i+135]:
        sum+=1
print(f'정확도 : {round(sum/len(predict)*100,2)}%')
```

```python
# accuracy_score 함수를 사용해서
#accuracy_score(테스트 데이터의 라벨과, 연산된 테스트 결과)를 매칭시켜봅니다
print(f'정확도 : {accuracy_score(y_test,predict).round(4)*100}%')
```

#### 추가로 markers의 값들이 어떤게 있는지 확인해 봅니다.

```python
# set the size
plt.rcParams['figure.figsize'] = [10, 8]

# remove ticks and values of axis
plt.xticks([])
plt.yticks([])

# markers' shape
all_shape=['.','o','v','^','>','<','s','p','*','h','H','D', 'd', '', '', '']
num = 0
for x in range(1, 5):
    for y in range(1, 5):
        num += 1        
        plt.plot(x, y, 
                 marker = all_shape[num-1], 
                 markerfacecolor='green', 
                 markersize=20, 
                 markeredgecolor='black')        
        plt.text(x+0.1, y, 
                 all_shape[num-1], 
                 horizontalalignment='left', 
                 size='medium', 
                 color='black', 
                 weight='semibold')     
plt.title('Markers', fontsize=20)
plt.show()
```

```python
# 사용할 마커들을 markers라는 이름으로 저장합니다.
markers=['H','D','*']

# 반복문을 사용해 ['setosa' 'versicolor' 'virginica'] 에서 벨류값들을 target변수에 저장합니다.
for i,target in enumerate(iris.target_names):
    # target_cluster변수에 데이터프레임의 모든 target컬럼의 데이터를 가져와 반복문의 target변수와 비교하여 저장시켜줍니다.
    target_cluster = df[df['target'] == target]
    
    # plt.scatter를 사용해  target_cluster의 결과값들을
    # x축엔 'petal length (cm)'
    # y축엔 'petal width (cm)'  을 넣어 데이터들을 비교해 봅니다.
    plt.scatter(x=target_cluster['petal length (cm)'], y=target_cluster['petal width (cm)'], marker=markers[i])
```
